```python
import autograd.numpy as np
import autograd
import matplotlib.pyplot as plt


N = 5
def w(x):
    W = np.zeros(N)
    for i in range(N):
        xi = x[i]
        if isinstance(x[i], autograd.numpy.numpy_boxes.ArrayBox):
            xi = x[i]._value
        W[i] = xi**2
    return W
        
def f(x):
    return np.dot(w(x), x)

g = autograd.grad(f)
x = np.random.rand(N)

print(f(x))
plt.scatter(x, w(x))
plt.title('w(x)')
plt.show()
print(g(x))
plt.plot(x, g(x))
plt.title('grad(f(x))')
plt.show()
```
