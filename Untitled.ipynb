```python
import autograd.numpy as np
import autograd
import matplotlib.pyplot as plt


N = 10
def w(x):
    W = np.zeros(N)
    for i in range(N):
        W[i] = x[i]**2
    return W
        
def f(x):
    return np.dot(w(x), x)

g = autograd.grad(f)
x = np.random.rand(N)

print(f(x))
plt.scatter(x, w(x))
plt.title('w(x)')
plt.show()
plt.plot(x, g(x))
plt.title('f(x)')
```
